# Cumulative Returns

```python
AAL['Cumulative Returns'] = (1+AAL['returns']).cumprod()
GOOG['Cumulative Returns'] = (1+GOOG['returns']).cumprod()
AMZN['Cumulative Returns'] = (1+AMZN['returns']).cumprod()
BHC['Cumulative Returns'] = (1+BHC['returns']).cumprod()
```

```python
AAL['Cumulative Returns'].plot(label ='AAL_CR',figsize=(15,7))
GOOG['Cumulative Returns'].plot(label ='GOOG_CR',figsize=(15,7))
AMZN['Cumulative Returns'].plot(label ='AMZN_CR',figsize=(15,7))
BHC['Cumulative Returns'].plot(label ='BHC_CR',figsize=(15,7))
plt.legend()
```

1) By observing the below line graph we can say that Amazon and google are the two best performing stocks.

2) By observing the cumulative returns we come to the conclusion that in all the 24 stocks 5 to 6 stockes are best as per the index SP500
